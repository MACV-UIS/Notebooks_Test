## <span style="color:red"> 1. Histograms as compact representation </span> 

Is a representation of the information that gives us an overall idea about the intensity distribution of an image. 
- It is just another way of understanding the image [1](https://docs.opencv.org/3.1.0/d1/db7/tutorial_py_histogram_begins.html)
- If we normalize the histograms between [0-1] we can obtain **the probability distribution of the image**

```python
I = cv.imread('../data/image_set/cougar_body/image_0007.jpg',0) # <==== Read as grayscale image
In = 255 - I
im_bins= 30

# from opencv
hist_cv = cv.calcHist([I],[0],None,[im_bins],[0,256])
bins_cv = np.array(np.linspace(0, 255, num=im_bins))
# from numpy
hist_np ,bins_np = np.histogram(I.ravel(),im_bins,[0,256])
print hist_np.shape ,bins_np.shape

plt.figure(figsize=(7,10))
plt.subplot(121), plt.imshow(I,'gray'), plt.title('Original') 
plt.subplot(122), plt.imshow(In,'gray'), plt.title('Negative') 
plt.figure(figsize=(15,3))
plt.subplot(141), plt.hist(I.ravel(),bins=im_bins), plt.grid(True) 
plt.subplot(142), plt.hist(In.ravel(),bins=im_bins), plt.grid(True)
plt.subplot(143), plt.bar( bins_cv, hist_cv.reshape(hist_cv.shape[0]),alpha=0.7, width=5,)
plt.subplot(144), plt.bar( bins_np[:-1], hist_np.reshape(hist_np.shape[0]),alpha=0.7, width=5);
```

## <span style="color:red"> 1. Histograms for segmentation </span> 

For simple images (with a uniform background and a well defined foreground) we can use the histogram as PDF to split the object from the rest. Namely in such images there is two distributions well defined. 

```python
I_ele1 = cv.imread('../data/image_set/elephant/image_0020.jpg',0) 
im_bins =30
seg= I_ele1 > 150

plt.figure(figsize=(10,3))
plt.subplot(131), plt.imshow(I_ele1,'gray')
plt.subplot(132), plt.imshow(seg,'gray')
plt.subplot(133), plt.hist(I_ele1.ravel(),bins=im_bins), 
plt.axvline(x=150,color='r',linestyle='--'); plt.grid(True); 
```

<img src="../imgs/expression.png" style="width:700px;heigth:100px">
1 - Segment the image using the above $g(x,y)$ equation. This will produce two groups of pixels: $G_1$ consisting of all pixels with intensity values $> T$, and $G_2$ consisting of pixels with values $\leq T$.

2 - Compute the average intensity values $m_1$ and $m_2$ for the pixels in $G_1$ and $G_2$ respectively.

3 - Compute a new threshold value:
$T = \frac{1}{2} (m_1 + m_2)$

4 - Repeat Steps 2 through 4 until the difference between values of $T$ in successive iterations is smaller than a predefined parameter $\Delta T$.

```python
def thr_value(f, dist_err = 0.5):
    count = 0
    done = False 
    T = np.mean(f.flatten())
    print "initial T: ",T
    while not(done):
        count = count + 1
        # 1 -
        g = f > T
        # 2 -
        m1 = np.mean(f[g])
        m2 = np.mean(f[~ g])
        # 3 -
        Tnext = 0.5*(m1 + m2)
        # 4 -
        done = abs(T - Tnext) < dist_err
        T = Tnext 
    
    print "mean value: ", T, " iterations: ", count
    return T
```

```python
seg =  I_ele1> thr_value(I_ele1)
plt.figure(figsize=(15,4))
plt.subplot(121),plt.imshow(I_ele1, cmap='gray'),plt.title('Original')
plt.subplot(122),plt.imshow(seg, cmap='gray'),plt.title('Segmentation')
plt.show()
```

## <span style="color:red"> 3. Histograms to obtain better images </span> 

We can also use histograms to obtain better image visualization. For intance, with **histogram equalization**, we can modify the dynamic range and contrast of an image. Pixels are namely on a reduced range and the idea is to expand them to whole spectrum. 

<img src="../imgs/equal.png" style="width:200px;heigth:100px">

- It is computed the histogram as the (PDF) $P_x (i) = \frac{n_i}{n}$
- It is computed the CDF(i) = $\sum_{j}^{i} p_x(j)$
- It is scaled the CDF between [0 - 255]

```python
I_cat1 = cv.imread('../data/image_set/cougar_body/image_0007.jpg',0)
I_ele1 = cv.imread('../data/image_set/elephant/image_0049.jpg',0)

I_equ_cat = cv.equalizeHist(I_cat1)
I_equ_ele = cv.equalizeHist(I_ele1)



plt.figure(figsize=(17,7))
plt.subplot(241), plt.imshow(I_cat1,'gray')
plt.subplot(242), plt.imshow(I_equ_cat,'gray')
plt.subplot(243), plt.imshow(I_ele1,'gray')
plt.subplot(244), plt.imshow(I_equ_ele,'gray')
plt.subplot(245), plt.hist(I_cat1.ravel(),bins=im_bins), plt.grid(True); 
plt.subplot(246), plt.hist(I_equ_cat.ravel(),bins=im_bins), plt.grid(True); 
plt.subplot(247), plt.hist(I_ele1.ravel(),bins=im_bins), plt.grid(True); 
plt.subplot(248), plt.hist(I_equ_ele.ravel(),bins=im_bins), plt.grid(True); 
```
